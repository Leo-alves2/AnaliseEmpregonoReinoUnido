Precisamos converter o salário do hora em salário anual. Considerando que o Reino Unido considera 40 h de trabalho por semana e o ano possui 52 semanas. Sendo assim 40 x 52 = 2080 horas por anos

```python
# Vamos usar uma função para converte

def converte_para_salario_ano(salary):
    if 'Per Hour' in str(salary):
        hora_salario = re.findall(r'\d+\.?\d*', salary)
        if hora_salario:
            return float(hora_salario[0]) * 2080
    return salary
```

```python
# Aplicando a função na coluna salário

df['Salary'] = df['Salary'].apply(converte_para_salario_ano)
```

```python
# Vamos verificar quais foram estes dados perdidos

""" 
Serve para filtrar o DataFrame df, e retornar apenas as linhas que atendem a dois critérios combinados. Vamos detalhar:

Desmembrando a Lógica:
df['salario_medio'].isna()
→ Retorna True para linhas onde a nova coluna salario_medio está vazia (ou seja, NaN).

df['Salary'].isna()
→ Retorna True para linhas onde a coluna original Salary está vazia.

~df['Salary'].isna()
→ O til ~ é um operador lógico NOT.
Isso inverte o resultado de isna(), ou seja:

Retorna True quando a coluna Salary não está vazia.

&
→ Operador AND lógico, que exige que ambas as condições sejam verdadeiras.
"""

df[df['salario_medio'].isna() & ~df['Salary'].isna()]
```

```python
def extrair_media_da_string(valor):

    # Se for uma string, limpa e transforma em lista
    if isinstance(valor, str):
        # Remove colchetes, espaços extras e quebra por vírgula
        valor = re.sub(r'[\[\]]', '', valor)  # remove colchetes
        partes = [v.strip() for v in valor.split(',')]
    elif isinstance(valor, list):
        partes = valor
    elif isinstance(valor, (float, int)):
        return float(valor)
    else:
        return np.nan

    # Extrai apenas os números válidos
    numeros = []
    for item in partes:
        if isinstance(item, (int, float)):
            numeros.append(float(item))
        elif isinstance(item, str):
            item_limpo = item.strip().replace('£', '')
            if item_limpo.replace('.', '', 1).isdigit():
                numeros.append(float(item_limpo))

    return np.mean(numeros) if numeros else np.nan
```

```python
df.loc[indices_com_problema, 'salario_medio'] = df['Salary'].apply(extrair_media_da_string)
```

A Coluna 'Skills'

O problema aqui é que a coluna Skills contém uma string com várias habilidades separadas por vírgula, e queremos transformá-la em uma estrutura de colunas binárias (0 ou 1 para presença/ausência), que é o conceito de One-Hot Encoding multirrótulo (multi-label).

```python
"""
Uma forma mais direta e eficiente de aplicar o One-Hot Encoding no caso de uma coluna com strings separadas por vírgula e espaço
 """
habilidades_separadas = df['Skills'].str.get_dummies(sep=', ')
```

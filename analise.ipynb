A coluna "Salary" que deveria ser númericos contem simbolos e caracteres, como pretendemos mostrar um histograma com a distribuição dos salário, precisamos de uma coluna seja númerica. No momento temos salário por ano e por hora, além dos valores nulos, que vamos considerar a média.

```python
# Verificar somente a coluna Salary

df['Salary'].head()
```

```python
# Primeiro vamos limpar os sinais de libras e substituir o "K" por 1000

df['Salary'] = df['Salary'].str.replace('£', '').str.replace('K','000')
```

```python
# Verificando como ficou

df['Salary'].head()
```

```python
# Remover as palavras utilizando expressão regular

df['Salary'] = df['Salary'].str.replace(r'\((Employer|Glassdoor) est\.?\)', '', regex=True).str.strip()
```

```python
# Verificando como ficou

df['Salary'].head()
```

Precisamos converter o salário do hora em salário anual. Considerando que o Reino Unido considera 40 h de trabalho por semana e o ano possui 52 semanas. Sendo assim 40 x 52 = 2080 horas por anos
